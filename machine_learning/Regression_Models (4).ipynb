***Import Required Libraries***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv("/content/drive/MyDrive/Project_NN/Data/Cleaned.csv")

df.head()
```

```python
results_ml = []

X = df.drop(columns=["price_usd_log"])
Y = df["price_usd_log"]

# Train / Test split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

# Scale features for linear model
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, Y_train)
Y_pred = lr_model.predict(X_test_scaled)

# Evaluation
lr_mse = mean_squared_error(Y_test, Y_pred)
lr_rmse = np.sqrt(lr_mse)
lr_r2 = r2_score(Y_test, Y_pred)

print("MSE linear:", lr_mse)
print("RMSE linear:", lr_rmse)
print("R2 Score linear:", lr_r2)
print("")

results_ml.append([
    "Linear Regression",
    lr_mse,
    lr_rmse,
    lr_r2
])

# Decision Tree Regressor
tree = DecisionTreeRegressor(max_depth=10,min_samples_split=20,random_state=42)
tree.fit(X_train, Y_train)
y_pred = tree.predict(X_test)
mse = mean_squared_error(Y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, y_pred)
print("MSE tree:", mse)
print("Decision Tree RMSE:", rmse)
print("Decision Tree R2:", r2)
print("")

results_ml.append([
    "Decision Tree",
    mse,
    rmse,
    r2
])

rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=25,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, Y_train)
y_pred = rf.predict(X_test)
mse = mean_squared_error(Y_test, y_pred)
print("MSE RF:", mse)
print("RF RMSE:", np.sqrt(mse))
print("RF R2:", r2_score(Y_test, y_pred))
print("")

results_ml.append([
    "Random Forest",
    mse,
    np.sqrt(mse),
    r2_score(Y_test, y_pred)
])
```

```python
ml_results_df = pd.DataFrame(
    results_ml,
    columns=["Model", "MSE", "RMSE", "R2"]
)

print(ml_results_df)

ml_results_df.to_csv("/content/drive/MyDrive/Project_NN/Results_of_ML_NN/results_ml.csv", index=False)
```
