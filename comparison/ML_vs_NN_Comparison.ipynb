```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
ml = pd.read_csv("/content/drive/MyDrive/Project_NN/Results_of_ML_NN/results_ml.csv")
nn = pd.read_csv("/content/drive/MyDrive/Project_NN/Results_of_ML_NN/results_nn.csv")

df = pd.concat([ml, nn])
print(df)
```

```python
metrics = ["MSE", "R2"]

x = np.arange(len(df["Model"]))
width = 0.35

plt.figure(figsize=(10,6))

plt.bar(x - width/2, df["MSE"], width, label="MSE")
plt.bar(x + width/2, df["R2"], width, label="R2")

plt.xticks(x, df["Model"], rotation=45)
plt.title("Model Comparison (MSE vs R2)")
plt.legend()
plt.grid(axis='y')

plt.show()
```
